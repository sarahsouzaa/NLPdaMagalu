##Análise de Sentimento

A análise de sentimento é um campo da área de análise de dados que utiliza técnicas de processamento de linguagem natural e inteligência artificial para extrair e entender informações sobre opiniões e emoções presentes em textos não
estruturados.

```python
PromptFixo= '''
#Resposta do Prompt
A saída do prompt precisa ser idêntica a este exemplo: Resposta = "Sua resposta aqui"
A resposta precisa ser: Sentimento Positivo, Neutro ou Negativo.

#Questão
Qual o sentimento desta frase?

frase:
'''

frase = "Meu produto chegou atrasado."
Prompt = f"{PromptFixo}{frase}"

print(f"Prompt enviado: {Prompt}")
print("\n")
print("Envio do Prompt:", datetime.datetime.now().time())
print("-" * 50)

response = model.generate_content(Prompt)
print(response.text)
```

##Análise de Tópicos

A análise de tópicos é outra técnica de processamento de linguagem natural que identifica os principais tópicos dentro de um documento em texto.
Enquanto a análise de sentimento extrais emoções, a de tópicos busca descobrir temas subjacentes que se conectam em diferentes partes de um texto.

```python
PromptFixo= '''
#Resposta do Prompt
A saída do prompt deve ser um dicionário em Python seguindo esta ordem: Resposta = "Tópico identificado : Sua resposta aqui"

#Questão
Analise a frase e realize uma análise de tópicos.

frase:
'''

frase = "Realizei uma compra dia 01/06 do qual apesar do atraso, após ligar no SAC, o atendente resolveu meu problema e fiquei satisfeito com o produto."

Prompt = f"{PromptFixo}{frase}"

print(f"Prompt enviado: {Prompt}")
print("\n")
print("Envio do Prompt:", datetime.datetime.now().time())
print("-" * 50)

response = model.generate_content(Prompt)
print(response.text)
```

##Análise de Polarização de Opinião

Também conhecida como polarizaão ideológica, é um fenômeno social no qual a sociedade se divide em grupos opostos com visões e valores distintos sobre algum tema.
Essa divisão pode ter natureza política, religiosa, cultural, entre outros.

```python
PromptFixo= '''
#Resposta do Prompt
A saída do prompt deve ser um dicionário em Python seguindo esta ordem: Resposta = "Indícios de polarização : Identifique o tema (Social, Cultural, Religioso ou Político)"

#Questão
Analise a frase e realize uma análise de polarização.

frase:
'''

frase = "Não gostei do atendimento do SAC, se o rapaz fosse evangélico certamente seria mais educado."

Prompt = f"{PromptFixo}{frase}"

print(f"Prompt enviado: {Prompt}")
print("\n")
print("Envio do Prompt:", datetime.datetime.now().time())
print("-" * 50)

response = model.generate_content(Prompt)
print(response.text)
```

```python
Dic = {"Sentimento" : "Positivo",
       "Tópico" : "Atendimento",
       "Opinião" : "Não Houve",
       "Entidade" : "Atendimento"}
pd.DataFrame(Dic, index = [0])
```

```python
PromptFixo = '''
#Função
Você é um analista responsável por analisar os comentários dos clientes.

#Resposta
A saída do prompt deve ser exatamente igual a este exemplo em um dicionário em Python:
{"'Análise de Sentimento' : ", "'Detecção de Entidade' : ", "'Análise de Tópicos' : ", "'Análise de Polaridade de Opinião' : "},
As respostas devem seguir o seguinte formato:
- Para Análise de Sentimento: (Positivo, Negativo ou Neutro)
- Para Detecção de Entidade: (Entidade identificada nomeada na frase)
- Para Análise de Tópicos: (Principal Tópico)
- Para Análise de Polaridade de Opinião: (Político, Social, Cultural, Religioso, ou de outra natureza.)

Observação:
A resposta do dicionário deve conter apenas chave e 1 valor.

Exemplo de resposta ideal:
{"Análise de Sentimento" : "Neutro",
"Detecção de Entidade" : "Estabelecimento",
"Análise de Tópicos" : "Reclamação",
"Análise de Polaridade de Opinião" : "Político"}

Analise a frase e preencha o dicionário.

'''

frase = "O atendimento desta loja é ótimo."

Prompt = f'''{PromptFixo}{frase}'''

response = model.generate_content(Prompt)
print(response.text)
```

```python
#Loop para processar a base de dados

Acumulo = pd.DataFrame()

for Loop, Comentario in enumerate(BaseComentarios[0:10]):
  #print(Loop, Comentario)

  time.sleep(2)

  print(Loop)


  print("Envio do Prompt:", datetime.datetime.now().time())
  Prompt = f'''{PromptFixo}{Comentario}'''

  try:

    response = model.generate_content(Prompt)
    tabela = pd.DataFrame(eval(response.text), index = [0])
    tabela["comentário"] = Comentario
    tabela["Loop"] = Loop

    Acumulo = pd.concat([Acumulo, tabela])

  except:
    print(f"#Erro no Loop: {Loop}#")

  print("Retorno", datetime.datetime.now().time())
```
